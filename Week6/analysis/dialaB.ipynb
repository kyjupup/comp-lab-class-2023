```python
import numpy as np
import matplotlib.pyplot as plt
# 替换为您的COLVAR文件路径
colvar_file = '/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaB/COLVAR'


# 使用NumPy加载数据，从第六行开始，选择要绘制的三列数据
data = np.loadtxt(colvar_file, skiprows=5, usecols=(0, 1, 2))

# 获取时间和三列数据
time = data[:, 0]
column1 = data[:, 1]
column2 = data[:, 2]


# 创建一个新的图形
plt.figure(figsize=(12, 6))

# 绘制三列数据
plt.plot(time, column1, label='psi')
plt.plot(time, column2, label='phi')


# 添加标签和标题
plt.xlabel('Time')
plt.ylabel('Data')
plt.title('Data Time Series')

# 添加图例
plt.legend()

# 显示图形
plt.show()





```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the grid file
grid_file = '/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaB/ex1_metad_A.grid.dat'

# Assuming the grid file has a structure like this:
# x_coord y_coord z_coord energy
# 1.0 2.0 3.0 -10.0
# 1.0 2.0 4.0 -9.5
# ...

# Use NumPy to read the data from the grid file
data = np.loadtxt(grid_file)

# Extract the energy column (assuming it's in the fourth column, change index if needed)
energy = data[:, 2]

# Calculate the negative of the energy column
negative_energy = -energy

# Create a new figure for plotting
plt.figure(figsize=(8, 6))

# Plot the negative energy
plt.plot(negative_energy, label='Negative Energy')

# Add labels and title
plt.xlabel('Data Point')
plt.ylabel('Negative Energy')
plt.title('Negative Energy from Grid File')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
