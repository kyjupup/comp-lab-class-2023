```python
import numpy as np
import matplotlib.pyplot as plt

# Specify the path to your fes.dat file
fes_file = "/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaA/dynamic/fes.dat"
try:
    # Load data from the fes.dat file, skipping the first five rows (header rows)
    data = np.loadtxt(fes_file, skiprows=5)

    # Check if the data array has at least two columns
    if data.shape[1] < 2:
        raise ValueError("The data array does not contain at least two columns.")

    # Extract columns for phi and free energy
    phi = data[:, 0]
    free_energy = data[:, 1]

    # Create a line plot of free energy vs. phi
    plt.figure(figsize=(10, 6))
    plt.plot(phi, free_energy, lw=2)

    # Set plot labels and title
    plt.xlabel("Dihedral Phi (degrees)")
    plt.ylabel("Free Energy (kcal/mol)")
    plt.title("Free Energy vs. Dihedral Phi")

    # Show the plot
    plt.grid(True)
    plt.show()

except IOError as e:
    print(f"An error occurred while reading the file: {e}")
except ValueError as e:
    print(f"Invalid data format: {e}")
except IndexError as e:
    print(f"Index error: {e}")

 
   
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Specify the paths to your fes.dat files for different time points
fes_files = ["/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaA/dynamic/fes_1.dat", 
             "/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaA/dynamic/fes_3.dat", 
             "/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaA/dynamic/fes_10.dat", 
             "/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaA/dynamic/fes_20.dat", 
             "/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaA/dynamic/fes_50.dat", 
             "/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaA/dynamic/fes_100.dat"]
# Initialize a figure
plt.figure(figsize=(10, 6))
plt.title("Free Energy vs. Dihedral Phi for Different Time Points")

# Loop through each time point and plot the free energy profile
for time_ns, fes_file in zip([0.1, 0.3, 1, 2, 5, 10], fes_files):
    try:
        # Load data from the fes.dat file, skipping the first five rows (header rows)
        data = np.loadtxt(fes_file, skiprows=5)

        # Check if the data array has at least two columns
        if data.shape[1] < 2:
            raise ValueError("The data array does not contain at least two columns.")

        # Extract columns for phi and free energy
        phi = data[:, 0]
        free_energy = data[:, 1]

        # Calculate the minimum free energy value
        min_free_energy = min(free_energy)

        # Subtract the minimum free energy from all data points
        free_energy -= min_free_energy

        # Plot the free energy vs. phi for the current time point
        plt.plot(phi, free_energy, lw=2, label=f"{time_ns} ns")

    except IOError as e:
        print(f"An error occurred while reading the file: {e}")
    except ValueError as e:
        print(f"Invalid data format: {e}")
    except IndexError as e:
        print(f"Index error: {e}")

# Set plot labels and legend
plt.xlabel("Dihedral Phi (degrees)")
plt.ylabel("Free Energy (kcal/mol)")
plt.legend(title="Time", loc="upper right")

# Show the combined plot
plt.grid(True)
plt.show()


```
