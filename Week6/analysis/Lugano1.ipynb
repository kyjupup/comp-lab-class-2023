```python
import numpy as np
import matplotlib.pyplot as plt

# Specify the path to your COLVAR file
colvar_file = "/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaA/dynamic/dynamic_COLVAR_A"


# Load data from the COLVAR file
data = np.loadtxt(colvar_file)

# Extract columns for time and CV1
time = data[:, 0]
cv1 = data[:, 1]

# Convert time to nanoseconds (assuming your time data is in picoseconds)
time_ns = time / 1000.0

# Define the time range (in nanoseconds) to plot
start_time_ns = 0
end_time_ns = 2

# Find the indices of data points within the specified time range
indices_within_range = np.where((time_ns >= start_time_ns) & (time_ns <= end_time_ns))

# Extract data within the specified time range
time_range = time_ns[indices_within_range]
cv1_range = cv1[indices_within_range]

# Create a scatter plot (spot plot) of CV1 vs. time
plt.figure(figsize=(10, 6))
plt.scatter(time_range, cv1_range, c=cv1_range, cmap='viridis', marker='o', s=20)

# Set plot labels and title
plt.xlabel("Time (ns)")
plt.ylabel("CV1")
plt.title("Collective Variable 1 Behavior (First 2 ns)")
plt.colorbar(label="CV1 Value")

# Show the plot
plt.grid(True)
plt.show()








```

```python
# Specify the path to your HILLS file
hills_file = "/home/yk3019/comp-lab-class-2023/Week6/Inputs/dialaA/dynamic/bck.0.HILLS"
# Load data from the HILLS file
data = np.loadtxt(hills_file)

# Extract columns for time and Gaussian height
time = data[:, 0]
gaussian_height = data[:, 1]
gaussian_height = data[:, 3]

# Convert time to nanoseconds (assuming your time data is in picoseconds)
time_ns = time / 1000.0

# Create a scatter plot (spot plot) of Gaussian height vs. time
plt.figure(figsize=(10, 6))
plt.scatter(time_ns, gaussian_height, c=gaussian_height, cmap='viridis', marker='o', s=20)

# Set plot labels and title
plt.xlabel("Time (ns)")
plt.ylabel("Gaussian Height")
plt.title("Gaussian Height Decrease (Well-Tempered Metadynamics)")

# Add a colorbar to indicate Gaussian height values
plt.colorbar(label="Gaussian Height")

# Show the plot
plt.grid(True)
plt.show()

```
