```python

import numpy as np
import matplotlib.pyplot as plt

# Read data from the XVG file
data = np.genfromtxt('/home/yk3019/comp-lab-class-2023/Week3/Data/potential.xvg', skip_header=24)

time = data[:, 0]  # Assuming time is in the first column
epot_values = data[:, 1]  # Assuming potential energy (Epot) values are in the second column

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(time, epot_values, label='potential energy', color='blue')
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('Energy Minimization')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Annotate the plot with a trendline
start_index = 0  # Index where the trendline starts
end_index = len(epot_values)  # Index where the trendline ends
plt.plot(time[start_index:end_index], np.poly1d(np.polyfit(time[start_index:end_index], epot_values[start_index:end_index], 1))(time[start_index:end_index]), color='red', linestyle='--', label='Trendline')

plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from the XVG file
data = np.genfromtxt('/home/yk3019/comp-lab-class-2023/Week3/Data/temperature.xvg', skip_header=24)
time = data[:, 0]  # Assuming time is in the first column
temperature_values = data[:, 1]  # Assuming temperature values are in the second column

# Set the rolling window size
window_size = 10  # Adjust this as needed

# Calculate the rolling average manually
rolling_avg = np.zeros_like(temperature_values)
for i in range(len(temperature_values)):
    start_idx = max(0, i - window_size // 2)
    end_idx = min(len(temperature_values), i + window_size // 2 + 1)
    rolling_avg[i] = np.mean(temperature_values[start_idx:end_idx])

# Create the Temperature vs. Time Plot with Rolling Average
plt.figure(figsize=(10, 6))
plt.plot(time, temperature_values, label='Temperature', color='blue')
plt.plot(time, rolling_avg, label=f'Rolling Average (Window Size {window_size})', color='red', linestyle='--')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Temperature')
plt.legend()
plt.grid(True)
plt.show()
```
