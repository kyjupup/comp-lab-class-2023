```python
import numpy as np
import matplotlib.pyplot as plt

xvg_file = "Data/1hz3_T310.run.25000000.energy.xvg"
data = np.loadtxt(xvg_file, comments=['@', '#'])
time_ns = data[:, 0] * 1000  
num_columns = data.shape[1] -1 
plt.figure(figsize=(12, 8)) 

for i in range(num_columns):
    plt.subplot(num_columns, 1, i+1)
    plt.plot(time_ns, data[:, i + 1], label=f'Quantity {i + 1}')
    plt.xlabel('Time (ns)')
    plt.ylabel(f'Quantity {i + 1} Value')
    plt.title(f'Quantity {i + 1} vs. Time (ns)')
    plt.legend()
    plt.grid(True)

plt.tight_layout()  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xvg_file = "Data/1hz3_T310.run.25000000.energy.xvg"
data = np.loadtxt(xvg_file, comments=['@', '#'])


time_ns = data[:, 0] * 1000  
kinetic_energy = data[:, 1]  
temperature = data[:, 3]  

# Create a figure and a left (temperature) and right (kinetic energy) axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot temperature on the left axis (ax1)
ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('Temperature (K)', color='tab:blue')
ax1.plot(time_ns, temperature, color='tab:blue', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis on the right (kinetic energy)
ax2 = ax1.twinx()
ax2.set_ylabel('Kinetic Energy', color='tab:red')
ax2.plot(time_ns, kinetic_energy, color='tab:red', label='Kinetic Energy')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add legends for both lines
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

# Title and grid
plt.title('Temperature and Kinetic Energy vs. Time')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import mdtraj as md


ubiquitin_traj = md.load('Data/1UBQ_processed.pdb')

num_hbonds = md.baker_hubbard(ubiquitin_traj)

print(f"Total number of hydrogen bonds: {num_hbonds}")
```

```python
import mdtraj as md

traj = md.load('Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='Data/1hz3_T310.start.nowater.gro')

```

```python
end_to_end_distances = md.compute_distances(traj, [[0, traj.n_atoms - 1]])
radius_of_gyration = md.compute_rg(traj)
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(traj.time, end_to_end_distances, label='End-End Distance (nm)', color='blue')

ax.plot(traj.time, radius_of_gyration, label='Radius of Gyration (nm)', color='red')


ax.set_xlabel('Time (ps)')
ax.set_ylabel('Value (nm)')
ax.set_title('End-End Distance and Radius of Gyration vs. Time')
ax.legend()

plt.grid(True)
plt.show()
```

```python
# Create histograms
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(end_to_end_distances, bins=50, density=True, color='blue', alpha=0.7)
plt.xlabel('End-End Distance (nm)')
plt.ylabel('Normalized Frequency')
plt.title('Normalized Histogram of End-End Distance')

plt.subplot(1, 2, 2)
plt.hist(radius_of_gyration, bins=50, density=True, color='red', alpha=0.7)
plt.xlabel('Radius of Gyration (nm)')
plt.ylabel('Normalized Frequency')
plt.title('Normalized Histogram of Radius of Gyration')

plt.tight_layout()
plt.show()
```
