```python
import numpy as np
import matplotlib.pyplot as plt
import os

# 读取能量文件的函数
def read_energy_file(file_path):
    data = np.loadtxt(file_path)
    time = data[:, 0]
    potential_energy = data[:, 1]
    return time, potential_energy

# NVT能量文件路径
nvt_file_path = '../Outputs/NVT-1.ener'
# NPT能量文件路径
npt_file_path = '../Outputs/NPT-1.ener'

# 读取NVT和NPT能量文件
nvt_time, nvt_potential_energy = read_energy_file(nvt_file_path)
npt_time, npt_potential_energy = read_energy_file(npt_file_path)

# 绘制NVT势能与时间关系图
plt.figure(figsize=(10, 6))
plt.plot(nvt_time, nvt_potential_energy, label='NVT Potential Energy', color='blue')
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('NVT Potential Energy vs. Time')
plt.legend()
output_path_nvt = '../Figures/nvt_potential_energy_vs_time.png'
plt.savefig(output_path_nvt)
plt.show()
print(f'NVT图形已保存至: {output_path_nvt}')

# 绘制NPT势能与时间关系图
plt.figure(figsize=(10, 6))
plt.plot(npt_time, npt_potential_energy, label='NPT Potential Energy', color='orange')
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('NPT Potential Energy vs. Time')
plt.legend()
output_path_npt = '../Figures/npt_potential_energy_vs_time.png'
plt.savefig(output_path_npt)
plt.show()
print(f'NPT图形已保存至: {output_path_npt}')

```

```python
# QM/MM能量文件路径
qmmm_file_path = '../Outputs/monitor/MONITOR-1.ener'

# 读取QM/MM能量文件
qmmm_time, qmmm_potential_energy = read_energy_file(qmmm_file_path)

# 绘制QM/MM势能与时间关系图
plt.figure(figsize=(10, 6))
plt.plot(qmmm_time, qmmm_potential_energy, label='QM/MM Potential Energy', color='green')
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('QM/MM Potential Energy vs. Time')
plt.legend()
output_path_qmmm = '../Figures/qmmm_potential_energy_vs_time.png'
plt.savefig(output_path_qmmm)
plt.show()
print(f'QM/MM图形已保存至: {output_path_qmmm}')
```
