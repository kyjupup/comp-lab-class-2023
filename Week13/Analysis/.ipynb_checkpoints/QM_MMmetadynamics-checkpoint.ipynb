```python
import numpy as np
import matplotlib.pyplot as plt


def read_monitor_file(file_path):
    data = np.loadtxt(file_path)
    time = data[:, 0]
    reaction_coordinate = data[:, 1]
    return time, reaction_coordinate

monitor_file_path = '../Outputs/monitor/MONITOR-COLVAR.metadynLog'


time_monitor, reaction_coordinate = read_monitor_file(monitor_file_path)


plt.figure(figsize=(10, 6))
plt.plot(time_monitor, reaction_coordinate, label='Reaction Coordinate', color='red')
plt.xlabel('Time (ps)')
plt.ylabel('Reaction Coordinate')
plt.title('Reaction Coordinate vs. Time')



plt.legend()
output_path_reaction_coord = '../Figures/reaction_coordinate_vs_time.png'
plt.savefig(output_path_reaction_coord)
plt.show()
print(f'Reaction Coordinate图形已保存至: {output_path_reaction_coord}')
```

```python
import numpy as np
import matplotlib.pyplot as plt


def read_monitor_file(file_path):
    data = np.loadtxt(file_path)
    time = data[:, 0]
    reaction_coordinate = data[:, 1]
    return time, reaction_coordinate


monitor_file_path = '../Outputs/monitor/MONITOR-COLVAR.metadynLog'


time_monitor, reaction_coordinate = read_monitor_file(monitor_file_path)


plt.figure(figsize=(10, 6))
plt.plot(time_monitor, reaction_coordinate, label='Reaction Coordinate', color='red')
plt.xlabel('Time (ps)')
plt.ylabel('Reaction Coordinate')
plt.title('Reaction Coordinate vs. Time')

plt.xticks(np.arange(0, max(time_monitor)+1, step=1))

plt.yticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['0', 'π/2', 'π', '3π/2', '2π'])

plt.legend()
output_path_reaction_coord = '../Figures/reaction_coordinate_vs_time(ylable.png'
plt.savefig(output_path_reaction_coord)
plt.show()
print(f'Reaction Coordinate图形已保存至: {output_path_reaction_coord}')
```

```python
import numpy as np

def read_monitor_file(file_path):
    data = np.loadtxt(file_path)
    time = data[:, 0]
    reaction_coordinate = data[:, 1]
    return time, reaction_coordinate



init_monitor_file_path = '../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog'


time_monitor, reaction_coordinate = read_monitor_file(monitor_file_path)


init_monitor_data = np.loadtxt(init_monitor_file_path)
time_init_monitor = init_monitor_data[:, 0]
reaction_coordinate_init = init_monitor_data[:, 1]


print('Metadynamics Reaction Coordinates:')
for time, coord in zip(time_monitor, reaction_coordinate):
    print(f'Time: {time} ps, Reaction Coordinate: {coord}')

print('\nInitial Simulation Reaction Coordinates:')
for time, coord in zip(time_init_monitor, reaction_coordinate_init):
    print(f'Time: {time} ps, Reaction Coordinate: {coord}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def read_monitor_file(file_path):
    data = np.loadtxt(file_path)
    time = data[:, 0]
    reaction_coordinate = data[:, 1]
    return time, reaction_coordinate

monitor_file_path = '../Outputs/monitor/MONITOR-COLVAR.metadynLog'

init_monitor_file_path = '../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog'


time_monitor, reaction_coordinate = read_monitor_file(monitor_file_path)


init_monitor_data = np.loadtxt(init_monitor_file_path)
time_init_monitor = init_monitor_data[:, 0]
reaction_coordinate_init = init_monitor_data[:, 1]

plt.figure(figsize=(10, 6))
plt.plot(time_monitor, reaction_coordinate, label='Reaction Coordinate (Metadynamics)', color='red')
plt.plot(time_init_monitor, reaction_coordinate_init, label='Reaction Coordinate (Init)', color='blue', linestyle='dashed')
plt.xlabel('Time (ps)')
plt.ylabel('Reaction Coordinate')
plt.title('Reaction Coordinate vs. Time')

plt.legend()
output_path_reaction_coord = '../Figures/reaction_coordinate_vs_time_initmonitor.png'
plt.savefig(output_path_reaction_coord)
plt.show()
print(f'Reaction Coordinate图形已保存至: {output_path_reaction_coord}')
```
